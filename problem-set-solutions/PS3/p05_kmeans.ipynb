### Implement K-means

```python
import numpy as np

A = imread('data/peppers-small.tiff')
K = 16
MAX_ITER = 300
N = A.shape[0] * A.shape[1]

A = A.reshape(-1, 3)

class_centers = A[np.random.randint(N, size=K)]
class_assignment = np.zeros(N) - 1

current_iter = 0
pre_cost = None
while current_iter < MAX_ITER:
    current_iter += 1

    # Assign class for each pixel
    pairwise_distance = np.linalg.norm(A[:, None, :] - class_centers[None, :, :], axis=-1)
    assignment = np.argmin(pairwise_distance, axis=1)
    if np.all(assignment == class_assignment):
        print("K-means converged in", current_iter, "iteractions.")
        break
    class_assignment = assignment
    
    # Update the center of each class
    class_centers = np.array([A[class_assignment == k].mean(axis = 0) for k in range(K)])
```

### Display compressed image

```python
P = imread("data/peppers-large.tiff")
f, axarr = plt.subplots(1, 2, sharey=True, figsize=(15,15))
axarr[0].imshow(P)
axarr[0].set_title('Original image')
P = P.reshape(512 * 512, 3)
B = np.zeros((512*512, 3))
pairwise_distance = np.linalg.norm(P[:, None, :] - class_centers[None, :, :], axis=-1)
assignment = np.argmin(pairwise_distance, axis=1)
for k in range(K):
    B[assignment == k] = class_centers[k]
B = B.reshape(512, 512, 3)
B = B.astype(int)
axarr[1].imshow(B)
axarr[1].set_title('Compressed image')
plt.show()
```
